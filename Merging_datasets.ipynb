```python
excel_file = 'total_waste_collection.xlsx' # name of Excel file
csv_file = 'total_waste_collection.csv'  # Choose a name for your CSV file

# Read the Excel file
df = pd.read_excel(excel_file)

# Save as CSV
df.to_csv('total_waste_collection.csv', index=False)
```

```python
df_total_waste = pd.read_csv('total_waste_collection.csv')
df_total_waste.rename(columns = {'Tota_ mixed_ waste': 'Total_mixed_waste', 'Total_waste_without_ CDW_and_other' : 'Total_waste_no_CDW_and_other' }, inplace=True )
df_total_waste.head()
```

```python
df_total_waste['Period'] = pd.to_datetime(df_total_waste['Period'], format='%Y')
```

```python
merged_df = pd.merge(
    df_tourists,
    df_GDP_capita,
    on=['Period', 'Autonomous_community'],
    how='outer'  # Use 'outer' to keep all data, even if it's missing in one DataFrame
)

# 2. Merge the remaining DataFrames one by one
merged_df = pd.merge(merged_df, df_unemployment, on=['Period', 'Autonomous_community'], how='outer')
merged_df = pd.merge(merged_df, df_HPI, on=['Period', 'Autonomous_community'], how='outer')
merged_df = pd.merge(merged_df, df_waste_capita, on=['Period', 'Autonomous_community'], how='outer')
merged_df = pd.merge(merged_df, df_total_waste, on=['Period', 'Autonomous_community'], how='outer')
merged_df

# Now 'merged_df' contains the combined data from all DataFrames
```
